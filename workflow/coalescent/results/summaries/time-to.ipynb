### Time since sweep ended

```python
tab=pd.read_csv('summary.tau.tsv',sep='\t')
Me='../../../../../auxillary/ne/bottleneck-1000G.ne'
Ne=read_Ne(Me)
out='summary.tau.time.tsv'

shp=tab.shape[0]
ar=[]
br=[]
cr=[]
dr=[]
for j in range(shp):
    sub=tab.iloc[j]
    st=sub.TRUE
    pt=0.5
    sl=sub.LOW
    sm=sub['AVG.1']
    su=sub.UPP
    q=0.05
    a=when_freq(q,st,pt,Ne,False)
    b=when_freq(q,sl,pt,Ne,False)
    c=when_freq(q,sm,pt,Ne,False)
    d=when_freq(q,su,pt,Ne,False)
    qr=[q for i in range(shp)]
    ar.append(a)
    br.append(b)
    cr.append(c)
    dr.append(d)
tab['AF1']=qr
tab['TRUETIME1']=ar
tab['LOWTIME1']=br
tab['MEDTIME1']=cr
tab['UPPTIME1']=dr

ar=[]
br=[]
cr=[]
dr=[]
for j in range(shp):
    sub=tab.iloc[j]
    st=sub.TRUE
    pt=0.5
    sl=sub.LOW
    sm=sub['AVG.1']
    su=sub.UPP
    q=0.01
    a=when_freq(q,st,pt,Ne,False)
    b=when_freq(q,sl,pt,Ne,False)
    c=when_freq(q,sm,pt,Ne,False)
    d=when_freq(q,su,pt,Ne,False)
    qr=[q for i in range(shp)]
    ar.append(a)
    br.append(b)
    cr.append(c)
    dr.append(d)
tab['AF2']=qr
tab['TRUETIME2']=ar
tab['LOWTIME2']=br
tab['MEDTIME2']=cr
tab['UPPTIME2']=dr

tab.to_csv(out,sep='\t',index=False)
```
