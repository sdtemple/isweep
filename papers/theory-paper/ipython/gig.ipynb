## Plotting the conditional coalescent time distribution ( | IBD segment length)

Versions for grayscale versus color are available.

Uncomment to save plots.

Neutral, positive selection at different frequencies, and demography are available.

Ne files must be available in right pointable folders.

```python
from isweep import *
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import scipy
```

#### Choose some core plot settings

```python
bins = 30
edgecolor = 'k'
xlim = (0,300)
ylim = (0,0.03)
rng = (0,300)
plt.rc('font',size=15)
```

```python
alpha = 0.5
cM = 2.
n = 5_000
```

## Constant Ne

```python
Ns = [10_000,25_000,100_000,1_000_000]
tit = 'Demography'
suffix = 'coaltime-v2-' + 'samplesize' + str(n) + '-cM' + str(cM)  + '-Ne-'
```

```python
cols = ['tab:' + i for i in ['blue','orange','green','red']]

for j in range(len(Ns)):
    N = Ns[j]
    col = cols[j]
    label = 'Ne = ' + str(N)
    simdata = simulate_ibd_constant(n,N,long_ibd=cM,short_ibd=cM)
    x = [t[3] for t in simdata[5]]
    
    plt.hist(x,
             bins,
             edgecolor=edgecolor,
             label=label,
             density=True,
             color=col,
             range=rng
            )
    
    plt.legend(loc='upper right',title=tit)
    
    plt.xlabel('Coalescent time')
    
    plt.ylabel('Density')
    
    plt.ylim(ylim)
    
    plt.xlim(xlim)

    plt.grid(alpha=alpha)

    plt.tight_layout()
    
    # plt.savefig('../figures/' + suffix + str(N) + '.tiff')
    # plt.savefig('../figures/' + suffix + str(N) + '.png')
    
    plt.show()
```

```python
cols = ['0.4','0.55','0.7','0.85']

for j in range(len(Ns)):
    N = Ns[j]
    col = cols[j]
    label = 'Ne = ' + str(N)
    simdata = simulate_ibd_constant(n,N,long_ibd=cM,short_ibd=cM)
    x = [t[3] for t in simdata[5]]
    
    plt.hist(x,
             bins,
             edgecolor=edgecolor,
             label=label,
             density=True,
             color=col,
             range=rng
            )
    
    plt.legend(loc='upper right',title=tit)
    
    plt.xlabel('Coalescent time')
    
    plt.ylabel('Density')
    
    plt.ylim(ylim)
    
    plt.xlim(xlim)

    plt.grid(alpha=alpha)
    
    # plt.savefig('../figures/' + suffix + str(N) + '-grayscale.tiff')
    # plt.savefig('../figures/' + suffix + str(N) + '-grayscale.png')
    
    plt.show()
```

## Demography
