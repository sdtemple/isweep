```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# increase font size if you want
plt.rc('font',size=12)
```

### Plotting all the data

```python
table = pd.read_csv('second.ranks.tsv.gz',sep='\t')
table.head()
```

```python
sns.scatterplot(table,
                x='POS',
                y='AAF',
                hue='ZDELTA',
                palette='Greys',
                edgecolor='k'
                )
plt.ylim(0,1)
plt.legend(frameon=True,
           shadow=True,
           loc='upper right',
           title='Score')
plt.ylabel('Allele frequency')
plt.xlabel('Base pair position')

plt.grid(alpha=0.25)

plt.title('Simulated sweep at location 4e6 and p(0)=0.68')
```

## Plotting a downsample of the data

```python
# set this parameter as you wish
# cuts this much data off
# from the left and right of the plot
bp_shrink_in = 5e5

x_left_end = table.POS.min() + bp_shrink_in
x_right_end = table.POS.max() - bp_shrink_in

# you could also manually set x_left_end and x_right_end
# instead of making it symmetric from the initial table

# set this parameter as you wish
# downsample to a fraction of the initial data
downsample_frac = 0.25
subtable = table.sample(frac=downsample_frac,replace=False)

# plot a vertical red line at a given bp
best_pos = 4e6
```

```python
sns.scatterplot(subtable,
                x='POS',
                y='AAF',
                hue='ZDELTA',
                palette='Greys',
                edgecolor='k'
                )
plt.ylim(0,1)
plt.legend(frameon=True,
           shadow=True,
           loc='upper right',
           title='Score')
plt.ylabel('Allele frequency')
plt.xlabel('Base pair position')

plt.xlim(x_left_end,x_right_end)

plt.axvline(best_pos,color='tab:red',linewidth=2,linestyle='dashed')

plt.grid(alpha=0.25)

plt.title('Simulated sweep at location 4e6 and p(0)=0.68')
```
